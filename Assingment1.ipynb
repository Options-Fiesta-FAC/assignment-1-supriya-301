```python
K = int(input("Enter the Strike Price: "))
P = int(input("Enter the Premium: "))
start, end = map(int, input("Enter the ").split())
S = range(start, end)
```

```python
data=[]
for s in S:
    row = {
        'S': s,
        'buy_call': max(0, s - K) - P,
        'sell_call': P - max(0, s - K),
        'buy_put': max(0, K - s) - P,
        'sell_put': P - max(0, K - s)
    }
    data.append(row)

import pandas as pd
df=pd.DataFrame(data)
print(df.head())
```

```python
import matplotlib.pyplot as plt

plt.scatter(S, df['buy_call'])
plt.xlabel("Spot Price")
plt.ylabel("Profit/loss")
plt.show()
```

```python
plt.scatter(S, df['sell_call'])
plt.xlabel("Spot Price")
plt.ylabel("Profit/loss")
plt.show()
```

```python
plt.scatter(S, df['buy_put'])
plt.xlabel("Spot Price")
plt.ylabel("Profit/loss")
plt.show()
```

```python
plt.scatter(S, df['sell_put'])
plt.xlabel("Spot Price")
plt.ylabel("Profit/loss")
plt.show()
```
